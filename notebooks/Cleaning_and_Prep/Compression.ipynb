```python
import numpy as np
import pandas as pd
from ucimlrepo import fetch_ucirepo 
```

```python
# fetch dataset 
online_retail = fetch_ucirepo(id=352) 
  
# data (as pandas dataframes) 
data = pd.concat([online_retail.data.features, online_retail.data.targets ], axis=1)
data
```

```python
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])

# data['Description'] = data[col].astype('category')
# data['CustomerID'] = data['CustomerID'].astype('category')
# data['Country'] = data['CustomerID'].astype('category')

data['Quantity'] = pd.to_numeric(data['Quantity'], downcast = 'integer')
data['UnitPrice'] = pd.to_numeric(data['UnitPrice'], downcast='float')
```

```python
data.to_parquet('../../data/online_retail.parquet', compression = 'zstd')
```
